```python
import os
import random
import shutil
```

```python


# 元のCSVファイルが含まれるフォルダのパス
source_folder = '/Users/u9004085802/Desktop/motion_detect_createML-master/data/0911_gyro_circle_increase_without_switch_but_including_post_data/train/circle'

# 移動先のフォルダのパス
destination_folder = '/Users/u9004085802/Desktop/motion_detect_createML-master/data/0911_gyro_circle_increase_without_switch_but_including_post_data/test/circle'

# フォルダ内のすべてのCSVファイルのリストを取得
csv_files = [f for f in os.listdir(source_folder) if f.endswith('.csv')]

# ランダムにファイルを選択して移動
num_files_to_move = int(len(csv_files) * 0.1)  # 10%を計算
files_to_move = random.sample(csv_files, num_files_to_move)

for file_to_move in files_to_move:
    source_file_path = os.path.join(source_folder, file_to_move)
    destination_file_path = os.path.join(destination_folder, file_to_move)
    
    # ファイルを移動
    shutil.move(source_file_path, destination_file_path)
    print(f"ファイル {file_to_move} を移動しました。")

print("移動が完了しました。")
```
