```python
import os
import pandas as pd
```

```python


# ソースフォルダとターゲットフォルダのパス
source_folder = '/Users/u9004085802/Desktop/motion_detect_createML-master/data/0911_gyro_circle_increase/train/circle'  # ソースフォルダのパス
target_folder = '/Users/u9004085802/Desktop/motion_detect_createML-master/data/0911_gyro_circle_increase/train/circle_reversed'  # ターゲットフォルダのパス

# ソースフォルダ内のすべてのCSVファイルを取得
csv_files = [f for f in os.listdir(source_folder) if f.endswith('.csv')]

# gyro_xカラムを反転させたCSVファイルを作成
for csv_file in csv_files:
    # CSVファイルのパス
    csv_file_path = os.path.join(source_folder, csv_file)
    
    # CSVファイルを読み込む
    df = pd.read_csv(csv_file_path)
    
    # gyro_xカラムを反転させる
    df['gyro_x'] = -df['gyro_x']
    
    # 新しいファイル名を生成
    new_file_name = os.path.splitext(csv_file)[0] + '_reversed.csv'
    
    # ターゲットフォルダに新しいCSVファイルを保存
    target_file_path = os.path.join(target_folder, new_file_name)
    df.to_csv(target_file_path, index=False)
    
    print(f"ファイル {csv_file} をコピーし、{new_file_name} として保存しました。")

print("処理が完了しました。")
```
